```python
df_1 = pd.read_csv('data/hist_data.csv')
plt.hist(df_1['aデータ'], alpha=0.5, bins=30, label='aデータ', color='orange')
plt.hist(df_1['bデータ'], alpha=0.5, bins=30, label='bデータ', color='cyan')
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
pd.read_csv('data/scatter_data.csv')
```

```python
df_2 = pd.read_csv('data/scatter_data.csv')
class0 = df_2[df_2['class'] == 0]
class1 = df_2[df_2['class'] == 1]
class2 = df_2[df_2['class'] == 2]
plt.scatter(class0['length (cm)'], class0['width (cm)'], label='class0', color='red')
plt.scatter(class1['length (cm)'], class1['width (cm)'], label='class1', color='blue')
plt.scatter(class2['length (cm)'], class2['width (cm)'], label='class2', color='green')
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.grid()
plt.legend()
plt.show()
```

```python
pd.read_csv('data/line_data.csv')
```

```python
df_3 = pd.read_csv('data/line_data.csv')
x = df_3['期間']
y = df_3['件数']
plt.plot(x, y)
plt.scatter(x, y)
plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.ylabel('件数')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
